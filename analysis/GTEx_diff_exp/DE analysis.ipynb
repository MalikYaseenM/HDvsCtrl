```python
%pylab inline
import pandas
norm_counts = pandas.read_csv('../../samples/GTEx/results/GTEX_salmon_norm.csv',index_col=0)
```

```python
ba9 = [_ for _ in norm_counts.columns if 'BA9' in _]
cau = [_ for _ in norm_counts.columns if 'CAU' in _]
ba9_median = norm_counts[ba9].median(axis=1)
cau_median = norm_counts[cau].median(axis=1)

scatter(log10(ba9_median),log10(cau_median),alpha=0.2)
our_fc = ba9_median/cau_median
```

```python
recount = pandas.read_csv('../../samples/GTEx/results/RSE_norm.csv',index_col=0)
recount = recount.loc[norm_counts.index]
ba9 = [_ for _ in recount.columns if 'BA9' in _]
cau = [_ for _ in recount.columns if 'CAU' in _]
```

```python
ba9_median = recount[ba9].median(axis=1)
cau_median = recount[cau].median(axis=1)
scatter(log10(ba9_median),log10(cau_median),alpha=0.1)
recount_fc = ba9_median/cau_median
```
